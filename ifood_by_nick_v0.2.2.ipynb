```python
count_num = 0
    
for offset in range(10000,12000,5):
    search_url = "https://ifoodie.tw/api/blog/?offset={}&limit=5&order_by=-date".format(offset)
    
    res = requests.get(search_url)
    soup = BeautifulSoup(res.text, 'lxml')
    
    pre_dict_list = soup.select('p')[0].text.split('response\": ')
    dict_list = pre_dict_list[1].split(''', "success"''')[0]
    
    petten = '\"description\": .+\.{3}\", '
    
    dict_list_without_description_list = re.split(petten, dict_list)
    
    dict_list_without_description = ''
    for entry  in dict_list_without_description_list:
        dict_list_without_description += entry
        
    try:
        json_objects = json.loads(dict_list_without_description)
    except:
        print('[ERROR]{}'.format(search_url))
        #continue
        
    with open('ifoodie_json_list_Debug{}_10000_12000.json'.format(count_num), 'a',encoding='utf8') as fa:
        data = json.dumps(json_objects, ensure_ascii=False)

        fa.write(data)
        print(search_url)
    
    count_num += 1
```
