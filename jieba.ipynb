```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
client = MongoClient()
db = client['test']
collect = db['test_ipeen']
test5 = collect.find({})
test5_list=[]
for c in test5:
    com = c['comment']
    
    com_cut = jieba.cut(com, cut_all=False)
    counts={}
    
    for word in com_cut:
        #print(word)
        if word in counts:
            counts[word]+=1
        else:
            counts[word]=1
    
    test5_list.append(counts)

#print(test5_list)    
    
```

```python
import re
```

```python
i = 0

for dicts in test5_list:
    i += 1
    vec = DictVectorizer()
    transformer=TfidfTransformer()#此方法會統計每個tf-idf詞語的權重  
    vec_count = vec.fit_transform(dicts) #fit_tr~是將字典轉為類似詞頻矩陣 (列,行) > 數量
    #print(vec_count)  #可用矩陣顯示.toarray()
    tfidf=transformer.fit_transform(vec_count)#fit_tr~是計算tf-idf 
    
    word=vec.get_feature_names()#取得該字典裡的所有字詞 
    
    weight=tfidf.toarray()#轉為tf-idf矩陣，元素xx[i][j]表示j詞在i這個字典中tf-idf的權重  
#     print(weight)
     
    mydict = {}
    print("-------這是第",i,"個字典詞語的tf-idf權重------")  
    for j in range(len(word)):  
        m = re.match(r'^[\u4E00-\u9FFFa-zA-Z]+$',word[j])
        
        if m is not None:
            
#             print(word[j],weight[0][j])
            mydict[word[j]] = weight[0][j]
    mydict1 = sorted(mydict.items(), key=lambda d:d[1], reverse = True)
    print(mydict1)         
```
